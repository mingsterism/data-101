```python
# importing basic libraries

import numpy as np

import pandas as pd

from pandas import Series, DataFrame

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./dummyDataWError.csv')
df
```

```python
# select only relevant columns
X = df[['a', 'b', 'c', 'd']]
y = df['target']
```

```python
# Standardizing the data  OPTIONAL
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
# We need to fit the scaler to our data before transformation
X = scaler.fit_transform(
    X.loc[:])
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
print("Train")
print(X_train)
print("test")
print(X_test)
print("Y train")
print(y_train)
print("Y test")
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error

y_pred = reg.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mae
# How to convert MAE to percentage accuracy
# https://scikit-learn.org/dev/modules/generated/sklearn.metrics.mean_absolute_percentage_error.html
```
