```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt 
pd.options.display.float_format = "{:,.2f}".format
plt.figure(figsize=(40,40)) 
pd.set_option('display.max_columns', None)
```

```python
rawDf = pd.read_csv('houses_to_rent_v2.csv')
rawDf
```

```python
# CHALLENGE: convert non numerical data to numerics
```

```python
# rename the columns to more managable names
df = rawDf.rename(
    columns={
        "hoa (R$)": "hoa", 
        "rent amount (R$)": "rentAmount", 
        "fire insurance (R$)": "fireInsurance", 
        "total (R$)": "total", 
        "parking spaces": "parkingSpaces", 
        "property tax (R$)": "propertyTax"})
df
```

```python
# Start scaling out the features and the target using standardscaler to ensure optimal performance of ML algorithms
features = ["area", "rooms", "bathroom", "parkingSpaces"]
x  = df.loc[:, features].values
x
```

```python
# standardize the features
standardX = StandardScaler().fit_transform(x)
standardX    # how does this work? 
```

```python
# seperating out the target
y = df.loc[:, 'furniture']
y
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(standardX)
principalDf = pd.DataFrame(data = principalComponents, columns = ['pc1', 'pc2'])
principalDf
```

```python
finalDf = pd.concat([principalDf, df[['furniture']]], axis=1)
finalDf
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)
ax.set_xlabel('pc1', fontsize=15)
ax.set_ylabel('pc2', fontsize=15)
ax.set_title('2 component pca', fontsize=20)
targets = ["furnished", "not furnished"]
color = ['r', 'b']
for target, color in zip(targets, color): 
    indicesToKeep = finalDf['furniture'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'pc1'], 
              finalDf.loc[indicesToKeep, 'pc2'], 
              c = color, 
              s = 50)
ax.legend(targets)
ax.grid()
```
